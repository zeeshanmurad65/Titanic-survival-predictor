### Data Preprocessing Steps

```python
# Removing those Columns which are not required such as name ,PassengerId , parch,sibsp ,Ticket
dataset.drop(['Name','PassengerId','Parch','SibSp','Ticket'] , axis=1 ,inplace=True)
```

```python
dataset.head(2)
```

```python
dataset.isnull().sum()
```

```python
(687/891)*100 # So Cabin more then 50 percent data is missing so remove this also
dataset.drop('Cabin' , axis=1 , inplace=True)
```

```python
# Filling Missing Values
dataset['Age'].fillna(dataset['Age'].mean() , inplace=True)
dataset['Embarked'].fillna(dataset['Embarked'].mode()[0] , inplace=True)
```

```python
# Label Encoding
from sklearn.preprocessing import LabelEncoder , StandardScaler
```

```python
dataset['Embarked'].unique()
```

```python
la=LabelEncoder()
```

```python
dataset['Embarked']=la.fit_transform(dataset['Embarked'])
dataset['Sex']=la.fit_transform(dataset['Sex'])
```

```python
dataset['Embarked'].unique()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.metrics import accuracy_score , classification_report , confusion_matrix
```

```python
input_data=dataset.drop('Survived' ,axis=1)
output=dataset['Survived']
```

```python
 X_train, X_test, y_train, y_test = train_test_split(input_data, output, test_size=0.23, random_state=42)
```

#### SVM Model

```python
Svm_pipeline=Pipeline([
    ('st' ,StandardScaler()),
    (('svm', SVC()))]
)
```

```python
Svm_pipeline.fit(X_train , y_train)
```

```python
svm_pred=Svm_pipeline.predict(X_test)
```

```python
accuracy_score(svm_pred,y_test)*100
```

#### Logistic Model

```python
lr_pipeline=Pipeline([
    ('st' ,StandardScaler()),
    ('lr',LogisticRegression(class_weight="balanced"))]
)
```

```python
lr_pipeline.fit(X_train , y_train)
```

```python
lr_pred=lr_pipeline.predict(X_test)
```

```python
accuracy_score(lr_pred,y_test)*100
```

#### Ramdom Forest Classifier

```python
Rnf_pipeline=Pipeline([
    ('st' ,StandardScaler()),
    ('lr',RandomForestClassifier(max_depth= 10,n_estimators=100))]
)
```

```python
Rnf_pipeline.fit(X_train , y_train)
```

```python
Rnf_pred=Rnf_pipeline.predict(X_test)
```

```python
accuracy_score(Rnf_pred,y_test)*100
```

```python
#### Gussian Navies Byes
```

```python
gnb_pipeline=Pipeline([
    ('st' ,StandardScaler()),
    ('lr',GaussianNB())]
)
```

```python
gnb_pipeline.fit(X_train,y_train)
```

```python
gnb_pred=gnb_pipeline.predict(X_test)
```

```python
accuracy_score(gnb_pred,y_test)*100
```
